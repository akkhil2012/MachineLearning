```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

import pandas

dataset = pandas.read_csv('voice.csv')



X,y = dataset.loc[:, dataset.columns != "label"],dataset.label
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1234,stratify=y)

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10],
              'gamma': [0.01, 0.1, 1, 10, 100]}

grid = GridSearchCV(SVC(), param_grid=param_grid, cv=15, verbose=3)
grid.fit(X_train, y_train)
print('Best score for SVC: {}'.format(grid.score(X_test, y_test)))
print('Best parameters for SVC: {}'.format(grid.best_params_))
```
